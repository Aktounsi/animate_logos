### 4. Insert label in concatenated dataframe
To-do: Insert label as new column in the following dataframe.

```python
label = pd.read_csv("data/label_animation_path/export_table_animation_path_label_9.4.2021.csv", skiprows=[1],
                                        dtype={"alias": object, "animation_file": object, "animation_id": int,
                                               "logo_id": int, "path_0_rating": object, "path_1_rating": object,
                                               "path_2_rating": object, "path_3_rating": object, 
                                               "path_4_rating": object, "path_5_rating": object, 
                                               "path_6_rating": object, "path_7_rating": object,
                                               "time": object})
```

```python
label_melt = pd.melt(label, id_vars=["alias","animation_file", "animation_id", "logo_id", "time"],
                                  value_vars=["path_0_rating", "path_1_rating", "path_2_rating", "path_3_rating",
                                              "path_4_rating", "path_5_rating", "path_6_rating", "path_7_rating"],
                                  var_name="path", value_name="label").assign(order_id=label_melt.path.str.extract("(\d+)")).astype({"order_id": 'int'})
```

```python
label_melt_id = label_melt.assign(order_id=label_melt.path.str.extract("(\d+)")).astype({"order_id": 'int'}).rename(columns={"animation_id":"animation_number"}).drop("path",1)
```

```python
label_melt_id[label_melt_id.logo_id == 376][label_melt_id.animation_number == 0]
```

```python
df_combined_id = df_combined.assign(
    logo_id=df_combined.file.str.extract("logo_(\d+)"),
    animation_number=df_combined.file.str.extract("animation_(\d+)")).astype({"logo_id": 'int',"animation_number": 'int'})
```

```python
df_combined_id
```

```python
# Combine dataframes and merge data
label_path_merged = pd.merge(df_combined_id, label_melt_id, how='left', on=["logo_id","animation_number","order_id"])
```

```python
# save dataframe
output = open(f"data/surrogate_model/animation_label.pkl", 'wb')
pickle.dump(label_path_merged, output)
output.close()
```
